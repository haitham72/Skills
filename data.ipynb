```python
import pandas as pd
import os

# Set the folder path
folder_path = r"Faz3"

# Get all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Process each CSV file and print shape information
lens = 0
for csv_file in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    
    try:
        # Try with UTF-8 first
        df = pd.read_csv(file_path, encoding='utf-8')
    except UnicodeDecodeError:
        try:
            # Fall back to latin-1 if UTF-8 fails
            df = pd.read_csv(file_path, encoding='latin-1')
        except UnicodeDecodeError:
            # Last resort: try cp1252
            df = pd.read_csv(file_path, encoding='cp1252')
    lens += 1
    
    # Extract the base name (remove the extension and comment part)
    base_name = csv_file.split('-comment--')[0]
    number = csv_file.split('-comment--')[1].split('.')[0]
    print(f"{base_name} - '{number}': '{df.shape[0]}'")

print("\nTotal comments files:", lens)
```
